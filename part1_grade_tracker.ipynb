## **Task 1 — Data Parsing & Profile Cleaning**

```python
# create the raw data
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
cleaned_students = []

for student in raw_students:
# name has leading/trailing whitespace removed and is converted to Title Case.
    # Step 1 : Clean name
    name = student["name"].strip().title()

# roll is converted from a string to an integer.
    # Convert roll to int
    roll = int(student["roll"])

# marks_str is split on ", " and each element is converted to an integer, producing a marks list.
    # Step 1: Extract the marks string from the student record
    marks_str = student["marks_str"]
    # Step 2: Split the string into a list of substrings
    split_values = marks_str.split(", ")
    # Step 3: Convert each substring to an integer using map
    int_values = map(int, split_values)
    # Step 4: Convert the iterator into a list
    marks = list(int_values)

#For each cleaned student, verify the name is valid: check that every word in the name contains only alphabetic characters. Print "✓ Valid name" or "✗ Invalid name" next to each student.
    # Step 1: Clean name
    name = student["name"].strip().title()
    # Step 2: Split name into words
    words = name.split()
    # Step 3: Validate each word (only alphabetic)
    is_valid = all(word.isalpha() for word in words)
    # Step 4: Assign validity message
    validity = "✓ Valid name" if is_valid else "✗ Invalid name"
    # Step 5: Print name with validity
    print(name, "-", validity)

# Print a formatted profile card for each cleaned student using f-strings:
    # Store cleaned data
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

    # Print profile card
    print("================================")
    print(f"Student : {name}  ({validity})")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")
    print("\n")


#  print the name in ALL CAPS and lowercase for the student with roll number 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("\nName transformations for roll 103:")
        print(student["name"].upper())
        print(student["name"].lower())
```

### **Task 2 — Marks Analysis Using Loops & Conditionals**

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# Using a for loop, print each subject alongside its marks and a grade label based on this scheme:

# Function to assign grade
def get_grade(score):
    if 90 <= score <= 100:
        return "A+"
    elif 80 <= score <= 89:
        return "A"
    elif 70 <= score <= 79:
        return "B"
    elif 60 <= score <= 69:
        return "C"
    else:
        return "F"

#  print Subject-wise output
print(f"Student: {student_name}\n")

for sub, mark in zip(subjects, marks):
    grade = get_grade(mark)
    print(f"{sub:10} : {mark} ({grade})")

# Calculations

print("\n--- Summary ---")
# total marks
total = sum(marks)
print(f"Total Marks        : {total}")

# Average marks
average = round(total / len(marks), 2)
print(f"Average Marks      : {average}")

# maximum marks
max_mark = max(marks)
max_index = marks.index(max_mark)
print(f"Highest Subject    : {subjects[max_index]} ({max_mark})")

# minimum marks
min_mark = min(marks)
min_index = marks.index(min_mark)
print(f"Lowest Subject     : {subjects[min_index]} ({min_mark})")


# --- Part 3: While loop for adding new subject and coresponding marks ---
new_count = 0

while True:
    subject_input = input("\nEnter subject name (or 'done' to stop): ").strip()

    if subject_input.lower() == "done":
        break
    # Subject input
    marks_input = input(f"Enter marks for {subject_input}: ").strip()

    # Validate marks input (should be integer and between 0 to 100)
    if not marks_input.isdigit():
        print("Invalid input: Marks must be a number.")
        continue

    marks_value = int(marks_input)

    if not (0 <= marks_value <= 100):
        print("Invalid input: Marks must be between 0 and 100.")
        continue

    # Add valid entry
    subjects.append(subject_input)
    marks.append(marks_value)
    new_count += 1

# print how many new subjects were added and the updated average across all subjects (original + new).---
updated_avg = round(sum(marks) / len(marks), 2)

print("\n--- Update Summary ---")
print(f"New subjects added : {new_count}")
print(f"Updated Average    : {updated_avg}")
```

## **Task 3 — Class Performance Summary**

```python
# Create list of students with their marks
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Assign the varialbles with default values
pass_count = 0
fail_count = 0
topper_name = ""
topper_avg = 0
total_avg_sum = 0

# Print table header
print("Name              | Average | Status")
print("----------------------------------------")

# Loop through each student using For loop
for name, marks in class_data:

    # Step 1: Calculate average marks
    avg = round(sum(marks) / len(marks), 2)

    # Step 2: Decide pass or fail
    status = "Pass" if avg >= 60 else "Fail"

    # Step 3: Count pass/fail students
    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1

    # Step 4: Check if current student is topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    # Step 5: Add to total average sum (for class average)
    total_avg_sum += avg

    # Step 6: Print formatted row
    print(f"{name:17} | {avg:7.2f} | {status}")

# Step 7: Calculate class average
class_avg = round(total_avg_sum / len(class_data), 2)

# Step 8: Print summary
print("\n--- Summary ---")
print(f"Passed Students : {pass_count}")
print(f"Failed Students : {fail_count}")
print(f"Class Topper    : {topper_name} ({topper_avg})")
print(f"Class Average   : {class_avg}")
```

## **Task 4 — String Manipulation Utility**

```python
# Original essay
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: Remove leading and trailing spaces
clean_essay = essay.strip()
print("Step 1 - Clean Essay:")
print(clean_essay)
print()

# Step 2: Convert to Title Case
title_case = clean_essay.title()
print("Step 2 - Title Case:")
print(title_case)
print()

# Step 3: Count occurrences of 'python' (case-insensitive)
# Since text is already lowercase, we can directly count
count_python = clean_essay.count("python")
print("Step 3 - Count of 'python':")
print(count_python)
print()

# Step 4: Replace 'python' with 'Python 🐍'
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Step 4 - Replace 'python':")
print(replaced_essay)
print()

# Step 5: Split essay into sentences using ". "
sentences = clean_essay.split(". ")
print("Step 5 - Sentences List:")
print(sentences)
print()

# Step 6: Print each sentence with numbering
print("Step 6 - Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):

    # Add '.' at the end if missing
    if not sentence.endswith("."):
        sentence += "."

    print(f"{i}. {sentence}")
```
